```python
#Commencons par cleaner les datas et garder les colonnes les plus pertinentes du dataset

df_mental_health=df_raw_data.drop(columns=['Timestamp','state','self_employed','care_options','wellness_program','coworkers',
       'supervisor', 'mental_health_interview', 'phys_health_interview','mental_vs_physical','benefits','seek_help', 
        'anonymity','mental_health_consequence', 'phys_health_consequence', 'comments'])
```

```python
df_mental_health.replace(('Yes', 'No'), (1, 0), inplace=True)
```

```python
df_mental_health
```

```python
df_mental_health["Gender"].unique()
```

```python
male_group = ['M', 'Male', 'male', 'm', 'Male-ish', 'maile','Cis Male', 
              'Mal', 'Male (CIS)',  'Make', 'Male ', 'Man', 'msle', 
              'Mail', 'cis male', 'Malr', 'Cis Man']
female_group = ['Female', 'Female ','female','Cis Female', 'F', 'Woman', 'f','woman', 
                'Femake', 'cis-female/femme', 'Female (cis)', 'femail']
other_group = ['Trans-female','non-binary', 'Nah', 'Enby', 'fluid', 'Genderqueer', 
               'Androgyne', 'Agender', 'Guy (-ish) ^_^', 'Neuter',
               'queer', 'queer/she/they','Trans woman',
               'Female (trans)','male leaning androgynous', 
               'ostensibly male, unsure what that really means','something kinda male?','A little about you','p','All']
```

```python
df_mental_health["Gender"].replace(male_group,"Male",inplace=True)
df_mental_health["Gender"].replace(female_group,"Female",inplace=True)
df_mental_health["Gender"].replace(other_group,"Others",inplace=True)
df_mental_health["Gender"].value_counts()
```

```python
#Existe-il une corrélation/un lien entre le fait de travailler dans le domaine des technologies et celui d'avoir recours à 
#un/des traitements pour les maladies mentales? Si ce lien existe, est-il fort?
```

```python
df_tech_and_treatment = df_mental_health.groupby(['tech_company'])[['treatment']].sum()
df_tech_and_treatment
#On remarque ici que sur un total de 632 employés ayant eu recours à un traitement pour traiter une/des maladies mentales,
#510 travaillent dans une compagnie du domaine de la technologie, contre seulement 122 qui ne travaillent pas pour une 
#compagnie de ce milieu professionnel. Ainsi, environ 81% des employés de compagnie dans le domaine de la technologie
#affirment avoir eu recours à un ou des traitement pour maladies mentales.
```

```python
#Y a-t-il des pays qui sont plus propices au développement de maladies mentales chez les employés en milieu des technologies?
```

```python
df_country = df_mental_health.groupby(['Country'])[['tech_company','treatment']].sum()

df_country['percentage'] = (df_country['treatment']/ df_country['tech_company'])*100

df_country.shape
```

```python
df_country.sort_values(by=['percentage'],ascending=False)

#On remarque que sur les 46 pays, 6 d'entres eux affichent un score de 100% en terme de corrélation entre le fait de 
#travailler pour une compagnie dans le domaine des technologies et le fait d'avoir recours à des traitements pour soigner
#des maladies mentales.Cependant, ces 6 pays ont tous moins de 5 employés ayant répondu à ce sondage, il est ainsi difficile
#d'établir une réelle corrélation entre ces variables. 

#Attardons nous maintenant à des pays comme les États-Unis ou le Canada. En effet, on remarque que ces deux pays d'Amérique
#du nord affichent un pourcentage d'environ 67%. Ce résultat est assez préoccupant. En soi, cela voudrait dire que près 
#de 2/3 des employés du milieu technologique ont recours à des traitements pour soigner des maladies mentales. 

#En contrepartie, des pays d'amérique du sud comme le Mexique et le Brésil n'affichent que la moitié du pourcentage de leurs 
#pays voisins du nord, c'est à dire 33%. On pourrait ici poser l'hypothèse que la température exerce un impact sur la santé 
#mentale des travaileurs,il peut être soit positif ou négatif. 
```

```python
#Est-ce que le fait de travailler à distance au moins 50% du temps pourrait avoir comme effet d'améliorer la santé mentale des
#employés du milieu technologique?
```

```python
df_remote = df_mental_health.groupby(['remote_work'])[['tech_company','treatment']].sum()
df_remote['percentage'] = (df_remote['treatment']/ df_remote['tech_company'])*100
df_remote

#On remarque ici qu'il y a 5% moins d'employés de compagnies du milieu technologique qui réclament un traitement pour soigner
#une maladie mentale lorsqu'ils travaillent 50% du temps ou plus à distance. On pourrait alors penser que le fait de travailler 
# à distance peut avoir un impact positif sur la santé mentale des employés. Or, encore une fois, ce pourcentage n'est pas 
#assez élevé pour pouvoir établir de fortes conclusions. 
```

```python
#Comment le nombre d'employés d'entreprises du milieu technologique peut impacter la santé mentale des employés?
```

```python
df_no_employee = df_mental_health.groupby(['no_employees'])[['tech_company','treatment']].sum()
df_no_employee['percentage'] = (df_no_employee['treatment']/ df_no_employee['tech_company'])*100
df_no_employee.sort_values(by=['percentage'], ascending=True)

#On remarque ici que le pourcentage d'employés des tech companies ayant recours à des traitements pour soigner les maladies
#mentales est plus faible lorsqu'il sagit de petites et moyennes entreprises que lorsqu'il s'agit de grosses entreprises de 
# 100, 500, 1000 employés et plus. Ainsi, on pourrait établir un lien entre ces variables et émettre l'hypothèse que dans de 
#plus petites compagnies, les employés sont mieux encadrés et plus écoutés. Dès lors, cela pourrait avoir un impact bénéfique
#sur leur santé mentale, car ils pourraient se sentir plus valorisés dans leur milieu professionnel et ainsi avoir moins 
#de chance de développer des troubles mentaux. 
```
